# Top 10 most costlier cars in dataset

```python
def format_values(value):
    if value>=10000000:
        return f"{value/10000000:.2f} Crores"
    elif value>=100000:
        return f"{value/100000:.2f} Lakhs"
    elif value>=1000:
        return f"{value/1000:.2f} K"
    else:
        return str(int(value))
    
```

```python
top_10_most_costlier_cars =df.groupby('car_name')['selling_price'].max().sort_values(ascending=False).head(10)
print(top_10_most_costlier_cars)
# Plot
plt.figure(figsize=(15,5))
plt.suptitle("Top 10 Most Costlier Cars in dataset", fontsize=30, fontstyle='italic')
ax=sns.barplot(x=top_10_most_costlier_cars.index, 
            y=top_10_most_costlier_cars.values, 
            palette="Set3", ec='black', linewidth=2.5)

formatted_labels = [format_values(i) for i in top_10_most_costlier_cars.values]

for bar, label in zip(ax.containers, formatted_labels):
    plt.bar_label(container=bar, labels=[label], label_type='edge')

plt.xticks(rotation=45)
plt.show()
```

```python
df_model = pd.read_csv("./data/cleaned_cars_dataset.csv")
df_model
```

# Feature Matrix => separating independent and dependent variables


```python
X = df_model.drop('selling_price',axis=1)
X
```

```python
y = df_model['selling_price']
y
```

```python
# train test split

from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0)
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
pd.DataFrame(X_test_scaled, columns= X.columns)
```

```python
# checking imp features
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X_train_scaled, y_train)
```

```python
imp_features = pd.Series(model.feature_importances_,index= X_train.columns)
imp_features.nlargest(15).plot(kind="barh", color="g")
```

# Model Building and Evaluation

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score, r2_score

```

```python
models = [LinearRegression, Ridge, Lasso, SVR, DecisionTreeRegressor, RandomForestRegressor]

mae = []
mse = []
rmse=[]
evs = []
r_squared_score = []

for model in models:
    regressor = model().fit(X_train_scaled, y_train)
    y_pred = regressor.predict(X_test_scaled)
    print(y_pred)

    mae.append(mean_absolute_error(y_true = y_test, y_pred=y_pred))
    mse.append(mean_squared_error(y_true = y_test, y_pred=y_pred))
    rmse.append(np.sqrt(mean_squared_error(y_true = y_test, y_pred=y_pred)))
    evs.append(explained_variance_score(y_true = y_test, y_pred=y_pred))
    r_squared_score.append(r2_score(y_true = y_test, y_pred=y_pred))
```

```python
ml_models_df = pd.DataFrame({
    "Models": ["LinearRegression", "Ridge", "Lasso", "SVR", "DecisionTreeRegressor", "RandomForestRegressor"],
    "mean_absolute_error":mae,
    "mean_squared_error":mse,
    "root_mean_squared_error":rmse,
    "explained_variance_score":evs,
    "r2_score":r_squared_score

})

ml_models_df.set_index("Models", inplace=True)
ml_models_df
```

# Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

params_grid = {
    LinearRegression:{},
    Ridge:{
        "alpha":[0.1,1.0,10.0]
    },
    Lasso:{
        "alpha":[0.001,0.01,0.1,1.0]
    },
    SVR:{
        "kernel":['rbf', 'linear'],
        "C":[0.1,1,10],
        'epsilon':[0.01,0.1,0.5]
    },
    DecisionTreeRegressor : {
        "max_depth":[None, 5,10,20],
        "min_samples_split":[2,5,10]
    },
    RandomForestRegressor:{
        "n_estimators":[50,100,200],
        "max_depth":[None, 5,10,20],
        "min_samples_split":[2,5,10]
    }
}

# metric list

models = [LinearRegression, Ridge, Lasso, SVR, DecisionTreeRegressor, RandomForestRegressor]

mae = []
mse = []
rmse=[]
evs = []
r_squared_score = []
best_models = []

for model in models:
    print(f"Tuning:{model.__name__}")
    if params_grid[model]:
        grid = GridSearchCV(model(), param_grid=params_grid[model], cv=5, scoring='r2')
        grid.fit(X_train_scaled,y_train)
        regressor = grid.best_estimator_
        print("Best Parameter:", grid.best_params_)
    else:
        regressor = model().fit(X_train_scaled, y_train)


    y_pred = regressor.predict(X_test_scaled)
    print(y_pred)

    mae.append(mean_absolute_error(y_true = y_test, y_pred=y_pred))
    mse.append(mean_squared_error(y_true = y_test, y_pred=y_pred))
    rmse.append(np.sqrt(mean_squared_error(y_true = y_test, y_pred=y_pred)))
    evs.append(explained_variance_score(y_true = y_test, y_pred=y_pred))
    r_squared_score.append(r2_score(y_true = y_test, y_pred=y_pred))
    best_models.append(regressor)
```

```python
ml_models_df = pd.DataFrame({
    "Models": ["LinearRegression", "Ridge", "Lasso", "SVR", "DecisionTreeRegressor", "RandomForestRegressor"],
    "mean_absolute_error":mae,
    "mean_squared_error":mse,
    "root_mean_squared_error":rmse,
    "explained_variance_score":evs,
    "r2_score":r_squared_score

})

ml_models_df.set_index("Models", inplace=True)
ml_models_df
```

```python
best_models
```

```python
import pickle
model_save_dir = "./saved_models/"
os.makedirs(model_save_dir, exist_ok=True)

for model in best_models:
    model_name = type(model).__name__
    model.fit(X_train_scaled, y_train)
    filename = f"{model_name}.pkl"
    print(f"Saving: {filename}")
    model_path = os.path.join(model_save_dir, filename)

    with open(model_path, 'wb') as f:
        pickle.dump(model, f)
    
    print(f"Trained and saved: {filename}")
```
